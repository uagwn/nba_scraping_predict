```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
player_stats = pd.read_csv("doncic.csv", index_col=0)
```

```python
player_stats
```

```python
player_stats.drop(['Rk', 'G', 'Date', 'Age', 'Unnamed: 7','GS','FG','FGA','FG%', '...', ' FT%', 'ORB', 'DRB', 'STL', 'BLK','FT','TOB', 'PF'], inplace=True, axis=1, errors='ignore')
```

```python
p_stats = player_stats.loc[player_stats['PTS'] != "Inactive"]
p_stats = p_stats[p_stats['PTS'] != 'Player Suspended']
p_stats = p_stats[p_stats['PTS'] != 'PTS']
p_stats = p_stats[p_stats['PTS'] != 'Did Not Dress']
p_stats = p_stats[p_stats['PTS'] != 'Did Not Play']
p_stats = p_stats[p_stats['PTS'] != 'Not With Team']
p_stats = p_stats[p_stats['PTS'] != 'TRB']
```

```python
p_stats
```

```python
p_stats['PTS'] = p_stats['PTS'].astype(float)
```

```python
p_stats['PTS'] = p_stats['PTS'].astype(int)
```

```python
p_stats
```

```python
p_stats.dtypes
```

```python
p_stats["target"] = (p_stats["PTS"] >= 30).astype("int")
```

```python
p_stats["opp_code"] = p_stats["Opp"].astype("category").cat.codes
```

```python
p_stats['away'] =  p_stats['Unnamed: 5'].isna()
```

```python
p_stats ['away'] = p_stats['away'].astype(int)
```

```python
p_stats['3ptbefore'] = p_stats["3P"].shift(1)
```

```python
p_stats ['day_before'] = p_stats["PTS"].shift(1)
p_stats['day_before'] = (p_stats['day_before'] < 25).astype("int")
```

```python
p_stats=p_stats.fillna(0)
```

```python
p_stats
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=20, min_samples_split=10, random_state=2)
```

```python
train = p_stats[0:40]
```

```python
test = p_stats[40:]
```

```python
predictors = ["opp_code", "away", "day_before","3ptbefore"]
```

```python
rf.fit(train[predictors], train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
p_acerto = accuracy_score(test["target"], preds)
```

```python
p_acerto
```
